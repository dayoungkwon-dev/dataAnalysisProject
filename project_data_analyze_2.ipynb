```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

from matplotlib import font_manager, rc
font_location='c:/Windows/fonts/malgun.ttf' #폰트 위치 확인
font_name=font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
```

```python
car_sales_rate_df_2023 = pd.read_csv('dataset/국내_자동차_판매순위_23년6월.csv', index_col=2)

car_sales_rate_df_2023
```

```python
car_sales_rate_df_2023.head(20).plot(kind='bar', ylabel='2023년 누계(6월 기준)')
```

```python
# 분류별로 비율을 더해서 파이그래프로 표현.
rate_df = car_sales_rate_df_2023.groupby("대분류")['비율'].sum().sort_values(ascending=False)
rate_df
```

```python
exp = [0.1, 0, 0, 0]
rate_df.plot(kind='pie',explode=exp, startangle=285, shadow=True,autopct='%.1f%%')
```
